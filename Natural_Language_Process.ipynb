```python

import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
STOPWORDS = stopwords
def _remove_stopwords(txt):
    
    words = txt.split()
    # words = txt.split(" ")
    for i, word in enumerate(words):
        if word in STOPWORDS:
            words[i] = " "
    return (" ".join(words))
```
